```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = np.object)
```

```python
X_num
```

```python
import pickle

# Normalizing data
transformer = Normalizer() 

transformer.fit(X_num) 


with open('transformer.pkl', 'wb') as file:
    pickle.dump(transformer, file) # this is how we save objects as a file

# loading object from my computer
with open('transformer.pkl', 'rb') as file:
    transformer = pickle.load(file)

x_normalized = transformer.transform(X_num)

x_normalized
```

### Processing Data

One Hot/Label Encoding (categorical)

```python
cat_data = pd.get_dummies(X_cat, drop_first=True)
```

Concat DataFrames

```python
data = pd.concat([x_normalizedDF,cat_data ], axis=1)
```

```python
data
```

```python
X=data
```

```python
y
```

### Linear Regression

#### Train-test split.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
# start the model
model = LinearRegression() # y = ai*xi + aii* xii + b

# train the model
model.fit(X_train,y_train) #I use the trining data set

model.coef_
```

```python
model.intercept_
```

Analyzing Results

```python
predictions  = model.predict(X_test) 
predictions.shape
```

```python
r2_score(y_test, predictions), mean_squared_error(y_test, predictions, squared=False), mean_squared_error(y_test, predictions)
```

R2:

```python
r2_score(y_test, predictions)
```

```python
#We have a good value of R2 since the closest is to 1 the better our model should predict 
```

RMSE:

```python
mean_squared_error(y_test, predictions, squared=False)
```

```python
# Lower values of RMSE indicate a better fit. We can see we have a kind a high RMSE 
```

MSE:

```python
mean_squared_error(y_test, predictions)
```

```python
# The MSE is the average of the square of the errors. The larger the number the larger the error. A MSE equal to 0 means the model is perfect. We can see we have a MSE of 142.43 which is high. 
```

MAE:

```python
mean_absolute_error(y_test, predictions)
```

```python
# measures the average magnitude of the errors in a set of predictions, the lower, the better. 
```
